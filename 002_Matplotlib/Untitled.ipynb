```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0.0, 0.5,0.02)
plt.plot(t,t,'r--',t,t**2,'gs',t,t**3,'b^')
plt.show()
```

```python
import numpy as np
data = {'a':np.arange(50),'d':np.random.randint(0,50,50),'c':np.random.randn(50)}
data['b'] =data['a'] + 10*np.random.randn(50)
data['d'] =np.abs(data['d']) *100
plt.scatter('a','b',c='c',s='d',data =data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
# student records using class
class  Student():
    name = ""  # class variable 
    roll = ""
    age  = ""
    def setting(self):
        print("*******************")
        print("setting name , roll  and age ")
        self.name = input("enter name : ")
        self.roll = input("enter roll : ")
        self.age = input("enter age   : ")
    def details(self):
        '''
        self ---> setting globalization of variablr
        self ---> object of class
        '''
        print("**********************")
        print("self : ",self)
        print("name : ",self.name)
        print("roll : ",self.roll)
        print("age  : ",self.age)
    def __del__(self): # destructor 
        print("self",self)
        print("delected")
std1 =  Student()
std2 =  Student()
std3 =  Student()
#*************************
std1.setting()
#**********************
std2.setting()
#***********************
std3.setting()
std1.details()
std2.details()
std3.details()
del std1
del std2
del std3
```

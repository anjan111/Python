```python
import math
print(math.sqrt(16))
print(math.factorial(5))
print(math.floor(14.5654354))
print(math.ceil(14.01))
print(math.pi)
```

```python
import math as m
print(m.sqrt(16))
print(m.factorial(5))
print(m.floor(14.5654354))
print(m.ceil(14.01))
print(m.pi)
```

```python
from math import sqrt,factorial,floor,ceil
print(sqrt(16))
print(factorial(5))
print(floor(14.5654354))
print(ceil(14.01))
print(pi)
```

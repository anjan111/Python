# Regular expression Special Characters 

---> Each and every character which having its own specific meaning

```python
# . ---> period character ( match any character(a -z / A-Z / 0-9/ special ) )
import re
s ='python Cython ckcython jython @ython 1ython gkdhk fkjgfjkg jkjdrkhgdkfjg jkkjh'
re.findall(r'.ython',s)
```

```python
re.search(r'^python',s).group() # its working like match method
```

```python
s='pyttthon pytthon python pyhon hskg bhdbj'
re.findall(r'pyt*hon',s) # 0 or more times repeat
```

```python
s ='pyttthon pytthon python pyhon hskg bhdbj'
```

```python
s='pyttthon pytthon python pyhon hskg bhdbj'
re.findall(r'pyt+hon',s) # 1 or more times repeat
```

```python
s='pyttthon pytthon python pyhon hskg bhdbj'
re.findall(r'pyt?hon',s) # 0 or 1 times repeat
```

```python
s='python pyxhon python pyxhon hskg bhdbj'
re.findall(r'py[tx]hon',s) 
```

```python
s='py1hon py2hon py3hon py4hon hskg bhdbj'
re.findall(r'py[1-5]hon',s) 
```

```python
s='pyttthon pytthon python pyhon hskg bhdbj'
re.findall(r'pyt{3}hon',s) 
```

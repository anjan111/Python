2. User defined functions

```python
'''
----->>> function defination

def  fun_name():----->> header of function
    Task
    ----        ----->> body of function
    ----
    
fun_name() # function call
fun_name() # function call
fun_name() # function call
'''
# waf sum of 2 numbers
def sum_of(): # function defination
    a = input("enter a : ")
    b = input("enter b : ")
    c = a+b
    print "sum : ",c
    
sum_of()# function call
print "**********"
sum_of()
```

```python
# waf for given number is prime number 

def isprime():
    num = input("enter num : ")
    for var in range(2,num):
        if(num % var == 0):
            break
    else:
        print "its prime number"
isprime()
```

```python
# arguments
'''
===>>> to send the data from function call to function defination
'''
def fun(a,b,c):# formal argu
    print a
    print b
    print c
    
var = 100
fun(123 , var, 12+12*34)# actaul arg
```

```python
# return 
'''
===>. sending the data  from fun def to fun call
===>. terminate the function
'''
def fun():
    print "hai"
    return 100
    print "bye"
a = fun()
print a
```

```python
# default  arguments --->>> 
'''
--->>>  formal arguments --->> copy of the actual arguments
-->>>  if your not not sending actual arg then it should be error
-->>> if your not not sending actual arg then formal arg can access default data

'''
def fun(a = None,b =None):
    print a
    print b
fun(10,20)
fun()
```

```python
# recursive functions
'''
--->> a fun it can calls itself
'''
def fun(i=1):
    print i, " : hello"
    fun(i+1)
fun()
```

```python
def fun(i=1):
    print i," : hello"
    if(i<10):
        fun(i+1)
fun()
```

```python
# wap print 1 to 10 using recursive fun
# wap print 10 to 1 using recursive fun
# wap print 1 to 10 and then 10 to 1 using recursive 
def fun(i =1):
    print i
    if(i < 10):
        fun(i+1)
    print i
fun()
```
